```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interactive

def viz(N):
    """Visualize Monte Carlo Integration.
    """
  
    
    # numbers from the domain of our function, f
    rng = np.random.RandomState(seed=1234)
    X = rng.uniform(a, b, size=N)
    area = ((b-a) * (1./N)) * f(X).sum()
    
    # optional - plot each point to fill-in the region
    for x in X:
        plt.axvline(x=x, ymax=f(x))
    plt.xlim(((np.pi/2), np.pi))
    plt.title("Area: {}".format(area))
    
# definite integral bounds and our function
a = np.pi/2
b = np.pi
f = np.sin

# select number of iterations, N
widget_N = widgets.IntText(value=1, min=1, step=2, max=100000, description="Iterations")
# run the visualization
interactive(viz, N=widget_N)
```

```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.exp(x) / (np.exp(x) + 1)
```

```python
# graph f solely for visualization
x_axis = np.linspace(-4, 4, 100)
plt.title("Sigmoid Function, $f$")
plt.xlim((-4, 4))
plt.plot(x_axis, f(x_axis))
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# define p whereby p is difficult to sample-from
p = stats.norm(1.5, 1.25)

# graph p solely for visualization
plt.hist(p.rvs(size=10000), bins=100, density=True, color='orange')
plt.xlim((-4, 4))
plt.title("True function, $p$")
plt.show()
```

```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

N = 100000

def visualize_is(mu, sigma):
    # generate values from q
    np.random.seed(1234)
    q = stats.norm(mu, sigma)
    X = q.rvs(size=N)
    variance = np.var(f(X) * p.pdf(X) / q.pdf(X))
    # graph p solely for visualization
    plt.hist(q.rvs(size=10000), bins=50, density=True, color='purple')
    plt.xlim((-4, 4))
    plt.title("$q$ variance is {}".format(variance))
    plt.show()

# define some UI widgets to represent parameter values
mus = widgets.FloatSlider(min=0.1, max=10, step=0.01)
sigmas = widgets.FloatSlider(min=0.1, max=10, step=0.01)

interact(visualize_is, mu=mus, sigma=sigmas)
```
